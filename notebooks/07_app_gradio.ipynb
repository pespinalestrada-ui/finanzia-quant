# 07 — App Gradio del proyecto

UI con 3 pestañas:
1. **Histórico** SAB.MC + eventos OPA marcados.
2. **Forecast 90d** con selector de modelo (Prophet/PyCaret/multi-horizonte/regresores).
3. **Narrativa GenAI** generada por el agente del NB 06.

La app real vive en `app/gradio_app.py`. Este notebook sirve para lanzarla desde Jupyter y para documentar el código.

## Lanzar la app

````python
import sys
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'app'))

# Carga del módulo de la app
import importlib.util
spec = importlib.util.spec_from_file_location('gradio_app', ROOT / 'app' / 'gradio_app.py')
ga = importlib.util.module_from_spec(spec)
spec.loader.exec_module(ga)

ga.app.launch(server_name='127.0.0.1', server_port=7860, share=False, inbrowser=False)
````

## Detener la app

````python
ga.app.close()
````
